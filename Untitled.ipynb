```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import os

# === Check TensorFlow version ===
print("TensorFlow version:", tf.__version__)

# === Load model ===
model_path = "path_to_your_model"  # e.g. "model.h5" or "saved_model/"
print(f"Loading model from: {model_path}")
model = load_model(model_path, compile=False)

# === Get model input shape ===
input_shape = model.input_shape  # e.g. (None, 224, 224, 3)
img_height, img_width = input_shape[1], input_shape[2]
print(f"Model expects input shape: {input_shape}")

# === Function to predict an image ===
def predict_image(img_path):
    if not os.path.exists(img_path):
        raise FileNotFoundError(f"Image not found: {img_path}")

    # Load and preprocess the image
    img = image.load_img(img_path, target_size=(img_height, img_width))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)  # (1, h, w, 3)
    img_array = img_array / 255.0  # normalize

    # Run prediction
    predictions = model.predict(img_array)

    print(f"\nPrediction for {os.path.basename(img_path)}:")
    print(predictions)

    # If model outputs class probabilities
    if predictions.shape[-1] > 1:
        predicted_class = np.argmax(predictions, axis=-1)[0]
        print(f"Predicted class index: {predicted_class}")
    else:
        print(f"Predicted value (binary/regression): {predictions[0][0]}")

# === Example usage ===
img_path = "path_to_your_image.jpg"
predict_image(img_path)
```
