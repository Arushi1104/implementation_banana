```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("ds_34_x_train.csv")   # can be .csv/.xlsx etc.

# View basic info
print(df.head())
print(df.describe())

# 1️⃣ Histogram for every numerical column
df.hist(figsize=(12, 8))
plt.suptitle("Histograms of Numerical Columns", fontsize=16)
plt.tight_layout()
plt.show()

# 2️⃣ Correlation Heatmap
plt.figure(figsize=(10, 6))
plt.imshow(df.corr(), interpolation='nearest')
plt.title("Correlation Heatmap")
plt.colorbar()
plt.xticks(range(len(df.columns)), df.columns, rotation=90)
plt.yticks(range(len(df.columns)), df.columns)
plt.tight_layout()
plt.show()

# 3️⃣ Line plot for each numerical column (if temporal data)
df.plot(figsize=(12, 6))
plt.title("Line Plot of Numerical Columns")
plt.xlabel("Index")
plt.ylabel("Values")
plt.legend()
plt.show()

# 4️⃣ Pairplot-style scatter matrix
pd.plotting.scatter_matrix(df, figsize=(12, 12))
plt.suptitle("Scatter Matrix", fontsize=16)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("ds_34_x_train.csv")

# ❗ Ignore the first column
df_num = df.iloc[:, 1:]   # everything except first column

# 1️⃣ Histogram
df_num.hist(figsize=(12, 8))
plt.tight_layout()
plt.savefig("histograms.png", dpi=200)
plt.close()

# 2️⃣ Correlation Heatmap with Labels
corr = df_num.corr()

plt.figure(figsize=(10, 6))
plt.imshow(corr, interpolation='nearest')
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Heatmap")
plt.tight_layout()
plt.savefig("correlation_heatmap.png", dpi=200)
plt.close()

# 3️⃣ Line Plot
df_num.plot(figsize=(12, 6))
plt.title("Line Plot of Numerical Columns (excluding first column)")
plt.savefig("line_plot.png", dpi=200)
plt.close()

print("All files saved (excluding first column).")
```
