```python
import joblib
import pandas as pd
import os
```

```python
# === MODEL PATH ===
model_path = "all_models/gradient_boost_model_s.pkl"

# === VERIFY MODEL EXISTS ===
if not os.path.exists(model_path):
    raise FileNotFoundError(f"Model file not found at {model_path}")

# === LOAD MODEL ===
print(f"Loading model from: {model_path}")
model = joblib.load(model_path)

# === FEATURE NAMES (must match training order) ===
features = [
    "Temp-int", "Press-int", "Humid-int",
    "Temp-ext", "Press-ext", "Humid-ext",
    "TGS20", "TGS02", "SGP",
    "SpA410", "SpB435", "SpC460", "SpD485", "SpE510", "SpF535", "SpG560", "SpH585",
    "SpR610", "SpI645", "SpS680", "SpJ705", "SpT730", "SpU760", "SpV810", "SpW860",
    "SpK900", "SpL940"
]
```

```python
print(f"Model expects {len(features)} comma-separated values.")
print("Press ENTER without typing anything to exit.\n")

while True:
    line = input("Enter values (comma-separated): ").strip()
    if line == "":
        print("Exiting.")
        break

    try:
        values = [float(x.strip()) for x in line.split(",")]
        if len(values) != len(features):
            print(f"Expected {len(features)} values, but got {len(values)}. Please try again.\n")
            continue
    except ValueError:
        print("Error: Non-numeric value detected. Please enter valid numeric values.\n")
        continue

    # Create DataFrame and predict
    df = pd.DataFrame([values], columns=features)
    pred = model.predict(df)[0]

    print("\nPrediction Result:")
    print(f"Predicted value: {pred}\n")

    again = input("Enter another line? (y/n): ").lower()
    if again != "y":
        break
```

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import os

# === Check TensorFlow version ===
print("TensorFlow version:", tf.__version__)

# === Load model ===
model_path = "path_to_your_model"  # e.g. "model.h5" or "saved_model/"
print(f"Loading model from: {model_path}")
model = load_model(model_path, compile=False)

# === Get model input shape ===
input_shape = model.input_shape  # e.g. (None, 224, 224, 3)
img_height, img_width = input_shape[1], input_shape[2]
print(f"Model expects input shape: {input_shape}")

# === Function to predict an image ===
def predict_image(img_path):
    if not os.path.exists(img_path):
        raise FileNotFoundError(f"Image not found: {img_path}")

    # Load and preprocess the image
    img = image.load_img(img_path, target_size=(img_height, img_width))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)  # (1, h, w, 3)
    img_array = img_array / 255.0  # normalize

    # Run prediction
    predictions = model.predict(img_array)

    print(f"\nPrediction for {os.path.basename(img_path)}:")
    print(predictions)

    # If model outputs class probabilities
    if predictions.shape[-1] > 1:
        predicted_class = np.argmax(predictions, axis=-1)[0]
        print(f"Predicted class index: {predicted_class}")
    else:
        print(f"Predicted value (binary/regression): {predictions[0][0]}")

# === Example usage ===
img_path = "path_to_your_image.jpg"
predict_image(img_path)
```
